```python
#Social distancing APP
import numpy as np
import cv2
import os
```

```python
model = "MobileNetSSD_deploy.caffemodel"
prototxt = "MobileNetSSD_deploy.prototxt.txt"

CLASSES = ["background", "aeroplane", "bicycle", "bird", "boat",
           "bottle", "bus", "car", "cat", "chair", "cow", "diningtable",
           "dog", "horse", "motorbike", "person", "pottedplant", "sheep",
           "sofa", "train", "tvmonitor"]
#COLORS = np.random.uniform(0, 255, size=(len(CLASSES), 3))
net = cv2.dnn.readNetFromCaffe(prototxt, model)
threshold = 0.4
min_distance = 50
```

```python
def create(frame,points,min_distance,prob,bounding_,out):
    violation = []
    if len(points)>=2:
        for i in range(len(points)):
            for j in range(i+1,len(points)):
                x = points[i][0] -  points[j][0]
                y = points[i][1] -  points[j][1]
                dis = x**2 + y**2
                dis = np.sqrt(dis)
                if dis < min_distance:
                    violation.append(i)
                    violation.append(j)
    color = [0,255,0]
    for k in range(len(points)):
        if k in violation:
            color = [0,0,255]
            
        confidence = prob[k]
        startX, startY, endX, endY = bounding_[k]
        label = "{}: {:.2f}%".format("Person", confidence * 100)
        print("[INFO] {}".format(label))
        cv2.rectangle(frame, (startX, startY), (endX, endY),
            color, 2)
        y = startY - 15 if startY - 15 > 15 else startY + 15
        cv2.putText(frame, label, (startX, y),
            cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)
        cv2.imshow('frame',frame)
        out.write(frame)
```

```python

path = "C:\\Users\\Asus\\Desktop\\CV notebooks"
cap = cv2.VideoCapture(os.path.join(path,"C:\\Users\\Asus\\Desktop\\CV notebooks\\Object_detection_cafee\\test.mp4"))

if(cap.isOpened()==False):
    print("Error: input not detected")

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
out = cv2.VideoWriter('outpy.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))

while(True):
    ret, frame = cap.read()
    frame = cv2.resize(frame,(300,300))
    h,w = frame.shape[:2]
    blob = cv2.dnn.blobFromImage(frame, 0.007843, (300, 300), 127.5)
    
    net.setInput(blob)
    detections = net.forward()
    points = []
    prob = []
    bounding_boxes = []
    for i in np.arange(0,detections.shape[2]):
        confidence = detections[0,0,i,2]
        if confidence>threshold:
            idx = int(detections[0,0,i,1])
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (startX, startY, endX, endY) = box.astype("int")
            
            if idx==15:
                points.append([(endX+startX)/2,(endY+startY)/2])
                prob.append(confidence)
                bounding_boxes.append([startX, startY, endX, endY])
            
            #This part tomorrow implement the distance function in open CV
            #red----unsafe/green----safe
            #####################################################
            
    create(frame,points,min_distance,prob,bounding_boxes,out)       
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

out.release()        
cap.release()
cv2.destroyAllWindows()
```
